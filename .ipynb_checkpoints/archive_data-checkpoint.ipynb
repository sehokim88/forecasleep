```python
# heart_rate 

parsed_hr_list = []
for date in dates:
    # later, replace sleep_end and sleep_start with actual corresponding values 
    sleep_end = datetime.strptime('06:00', '%H:%M').time()
    sleep_start = datetime.strptime('22:00', '%H:%M').time()
    # import
    filename = f'heart_rate-{date}.json'
    df = pd.read_json(f'{path}{filename}')
    # aggregate
    daytime_hr_dict = aggregate_hr(df, sleep_end, sleep_start)
    # append
    parsed_hr_list.append(daytime_hr_dict)
```

```python
daytime_hr = pd.DataFrame(parsed_hr_list)
daytime_hr
```

```python
# resting_heart_rate

# import
rhrs = pd.DataFrame()
for date in ['2017-11-14', '2018-11-14']:
    try:
        filename = f'resting_heart_rate-{date}.json'
        rhr = pd.read_json(f'{path}{filename}')
    except:
        continue

    rhrs = pd.concat([rhrs, rhr])
    
# expand
rhrs['resting_hr'] = rhrs['value'].apply(lambda x: x['value'])
rhrs['date'] = rhrs['dateTime'].apply(lambda x: x.date())

# trim
mask = (rhrs['date'] >= dates[0]) & (rhrs['date'] <= dates[-1])
rhrs = rhrs.loc[mask, ['date', 'resting_hr']].copy()

rhrs
```

## get active minute table 

```python
# sedentary_minute
sedentary_minute = pd.DataFrame()
for date in dates:
    try:
        filename = f'sedentary_minutes-{date}.json'
        sm = pd.read_json(f'{path}{filename}')
    except:
        continue

    sedentary_minute = pd.concat([sedentary_minute, sm])
    
sedentary_minute
```

```python
# lightly_active_minute
lightly_active_minute = pd.DataFrame()
for date in dates:
    try:
        filename = f'lightly_active_minutes-{date}.json'
        lam = pd.read_json(f'{path}{filename}')
    except:
        continue

    lightly_active_minute = pd.concat([lightly_active_minute, lam])
    
lightly_active_minute
```

```python
# moderately_active_minute
moderately_active_minute = pd.DataFrame()
for date in dates:
    try:
        filename = f'moderately_active_minutes-{date}.json'
        mam = pd.read_json(f'{path}{filename}')
    except:
        continue

    moderately_active_minute = pd.concat([moderately_active_minute, mam])
```

```python
# very_active_minute
very_active_minute = pd.DataFrame()
for date in dates:
    try:
        filename = f'very_active_minutes-{date}.json'
        vam = pd.read_json(f'{path}{filename}')
    except:
        continue

    very_active_minute = pd.concat([very_active_minute, vam])
    
very_active_minute
```

```python
# active_minutes

active_minutes = sedentary_minute\
.merge(lightly_active_minute, on='dateTime')\
.merge(moderately_active_minute, on='dateTime')\
.merge(very_active_minute, on='dateTime')


active_minutes.columns = ['date','sedentary', 'lightly', 'moderately', 'very']
active_minutes['total'] = active_minutes.sum(axis=1)
active_minutes['date'] = active_minutes['date'].apply(lambda x: x.date()).copy()
active_minutes
```

## get sleep table

```python
filename = 'sleep-2019-02-27.json'
sleep_raw = pd.read_json(f'{path}{filename}')
sleep_raw
```

```python
# sleep
filename = 'sleep-2018-05-03.json'
sleep_raw = pd.read_json(f'{path}{filename}')
sleep = pd.DataFrame()
def parse_time(time_str):
    return datetime.strptime(time_str, '%Y-%m-%dT%H:%M:%S.%f').time()
def parse_date(time_str):
    return datetime.strptime(time_str, '%Y-%m-%dT%H:%M:%S.%f').date()

sleep['date'] = sleep_raw['endTime'].apply(parse_date) - timedelta(days=1)
sleep['sleep_start'] = sleep_raw['startTime'].apply(parse_time)
sleep['sleep_end'] = sleep_raw['endTime'].apply(parse_time)
sleep['bed'] = sleep_raw['timeInBed']
sleep['sleep'] = sleep_raw['minutesAsleep']
sleep['awake'] = sleep_raw['minutesAwake']
try:
    sleep['deep'] = sleep_raw.levels.apply(lambda x: x['summary']['deep']['minutes'])
except:
    sleep['deep'] = np.nan
try:    
    sleep['rem'] = sleep_raw.levels.apply(lambda x: x['summary']['rem']['minutes'])
except:
    sleep['rem'] = np.nan
try:
    sleep['light'] = sleep_raw.levels.apply(lambda x: x['summary']['light']['minutes'])
except:
    sleep['light'] = np.nan
try:
    sleep['wake'] = sleep_raw.levels.apply(lambda x: x['summary']['wake']['count'])
except:
    sleep['wake'] = np.nan
sleep
```
