# Preprocessing & Pipelining

```python
# deep_fill
# preprocessing
sleep['deep_fill'] = sleep['deep']
# pipeline
deep_fill_fit = Pipeline([
    ('select', ColumnSelector(name='deep_fill')),
    ('fillna', StageFiller(X_total=sleep['sleep']))
])
# graph
```

```python
# start_time
# preprocessing
sleep['start_time'] = sleep['start'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %I:%M%p').time())
# pipeline
start_time_fit = Pipeline([
    ('select', ColumnSelector(name='start_time')),
    ('splittime', TimeScaler())
])
# graph
```

```python
# end_time
# preprocessing
sleep['end_time'] = sleep['end'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %I:%M%p').time())
# pipeline
end_time_fit = Pipeline([
    ('select', ColumnSelector(name='end_time')),
    ('splittime', TimeScaler())
])
# graph
```

```python
# Feature Union
union = FeatureUnion([
    ('intercept', Intercept()),
    ('start_time', start_time_fit),
    ('end_time', end_time_fit),
    ('deep_fill', deep_fill_fit)
])

union.fit(sleep)
transformed_data = union.transform(sleep)
```

# Modeling

```python
# model fit
y = transformed_data['deep_fill']
X = transformed_data.drop('deep_fill', axis=1)
lin = LinearRegression()
lin.fit(X, y)
```

```python
# interface for prediction
# start = input('time to go bed: ')
# end = input('time to wake up: ')
start = '22:00'
end = '6:00'
predictors = np.ones(X.shape[1])
predictors[1:] = input_time_scaler(start,end)
y_ = lin.predict(predictors.reshape(1,-1))[0]
print(f'predicted deep sleep minutes for tonight: {y_:4.2f}')
```

```python
# cross validation for scoring
kf = KFold(n_splits=5, shuffle=True)
scores = []
for train, test in kf.split(X):
    
    lin = LinearRegression()
    lin.fit(X.loc[train],y.loc[train])
    y_ = lin.predict(X.loc[test])
    mse = np.mean((y_-y.loc[test])**2)
    scores.append(mse)
    
np.mean(scores)
```

```python
# create new columns
sleep['start'] = pd.to_datetime(sleep.start)
sleep['end'] = pd.to_datetime(sleep.end)
sleep['date'] = sleep['start'].apply(lambda x: x.date())
sleep['asleep'] = sleep['start'].apply(lambda x: x.time())
sleep['awake'] = sleep['end'].apply(lambda x: x.time())

# calculate and add 'wake' 
wake = np.zeros(len(sleep.index))
wake[:-1] = pd.to_timedelta(sleep.start[:-1].values - sleep.end[1:].values)
sleep['wake'] = pd.to_timedelta(wake)

# remove observations where recording was skipped
unskipped_dates_mask = (sleep['wake'] < dt.timedelta(days=1)) & (sleep['wake'] > dt.timedelta(hours=1))
sleep_remove_skipped = sleep[unskipped_dates_mask].copy()

# time data to minutes
sleep_remove_skipped['wake'] = sleep_remove_skipped['wake'].apply(lambda x: x.seconds//60)

```
